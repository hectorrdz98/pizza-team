```python
import pandas as pd
```

```python
# Load of all .csv
df_ubereats_centro = pd.read_csv(r'csv/ubereats-pizza-centro.csv')
df_ubereats_alamos = pd.read_csv(r'csv/ubereats-pizza-alamos.csv')
df_ubereats_juriquilla = pd.read_csv(r'csv/ubereats-pizza-juriquilla.csv')
df_ubereats_milenio = pd.read_csv(r'csv/ubereats-pizza-milenio.csv')
df_ubereats_refugio = pd.read_csv(r'csv/ubereats-pizza-refugio.csv')
df_ubereats_balvanera = pd.read_csv(r'csv/ubereats-pizza-balvanera.csv')
df_ubereats_campanario = pd.read_csv(r'csv/ubereats-pizza-campanario.csv')
df_ubereats_cerritocolorado = pd.read_csv(r'csv/ubereats-pizza-cerritocolorado.csv')
df_ubereats_cimatario = pd.read_csv(r'csv/ubereats-pizza-cimatario.csv')
df_ubereats_el_pueblito = pd.read_csv(r'csv/ubereats-pizza-el-pueblito.csv')

df_rappi_centro = pd.read_csv(r'csv/rappi-pizza-centro.csv')
df_rappi_alamos = pd.read_csv(r'csv/rappi-pizza-alamos.csv')
df_rappi_juriquilla = pd.read_csv(r'csv/rappi-pizza-juriquilla.csv')
df_rappi_milenio = pd.read_csv(r'csv/rappi-pizza-milenio.csv')
df_rappi_refugio = pd.read_csv(r'csv/rappi-pizza-refugio.csv')
df_rappi_balvanera = pd.read_csv(r'csv/rappi-pizza-balvanera.csv')
df_rappi_campanario = pd.read_csv(r'csv/rappi-pizza-campanario.csv')
df_rappi_cerritocolorado = pd.read_csv(r'csv/rappi-pizza-cerritocolorado.csv')
df_rappi_cimatario = pd.read_csv(r'csv/rappi-pizza-cimatario.csv')
df_rappi_el_pueblito = pd.read_csv(r'csv/rappi-pizza-el-pueblito.csv')

# Get dataframes per colony
df_centro = pd.concat([df_ubereats_centro, df_rappi_centro], ignore_index=True)
df_alamos = pd.concat([df_ubereats_alamos, df_rappi_alamos], ignore_index=True)
df_juriquilla = pd.concat([df_ubereats_juriquilla, df_rappi_juriquilla], ignore_index=True)
df_milenio = pd.concat([df_ubereats_milenio, df_rappi_milenio], ignore_index=True)
df_refugio = pd.concat([df_ubereats_refugio, df_rappi_refugio], ignore_index=True)
df_balvanera = pd.concat([df_ubereats_balvanera, df_rappi_balvanera], ignore_index=True)
df_campanario = pd.concat([df_ubereats_campanario, df_rappi_campanario], ignore_index=True)
df_cerritocolorado = pd.concat([df_ubereats_cerritocolorado, df_rappi_cerritocolorado], ignore_index=True)
df_cimatario = pd.concat([df_ubereats_cimatario, df_rappi_cimatario], ignore_index=True)
df_pueblito = pd.concat([df_ubereats_el_pueblito, df_rappi_el_pueblito], ignore_index=True)
```

```python
# Remove MX$ from price-food
df_centro['price-food'] = df_centro['price-food'].apply(lambda x: (float(x[3:] if x[0] == 'M' else x[1:]) if not isinstance(x, int) else float(x)) if not isinstance(x, float) else x)
df_alamos['price-food'] = df_alamos['price-food'].apply(lambda x: (float(x[3:] if x[0] == 'M' else x[1:]) if not isinstance(x, int) else float(x)) if not isinstance(x, float) else x)
df_juriquilla['price-food'] = df_juriquilla['price-food'].apply(lambda x: (float(x[3:] if x[0] == 'M' else x[1:]) if not isinstance(x, int) else float(x)) if not isinstance(x, float) else x)
df_milenio['price-food'] = df_milenio['price-food'].apply(lambda x: (float(x[3:] if x[0] == 'M' else x[1:]) if not isinstance(x, int) else float(x)) if not isinstance(x, float) else x)
df_refugio['price-food'] =df_refugio['price-food'].apply(lambda x: (float(x[3:] if x[0] == 'M' else x[1:]) if not isinstance(x, int) else float(x)) if not isinstance(x, float) else x)
df_balvanera['price-food'] = df_balvanera['price-food'].apply(lambda x: (float(x[3:] if x[0] == 'M' else x[1:]) if not isinstance(x, int) else float(x)) if not isinstance(x, float) else x)
df_campanario['price-food'] = df_campanario['price-food'].apply(lambda x: (float(x[3:] if x[0] == 'M' else x[1:]) if not isinstance(x, int) else float(x)) if not isinstance(x, float) else x)
df_cerritocolorado['price-food'] = df_cerritocolorado['price-food'].apply(lambda x: (float(x[3:] if x[0] == 'M' else x[1:]) if not isinstance(x, int) else float(x)) if not isinstance(x, float) else x)
df_cimatario['price-food'] = df_cimatario['price-food'].apply(lambda x: (float(x[3:] if x[0] == 'M' else x[1:]) if not isinstance(x, int) else float(x)) if not isinstance(x, float) else x)
df_pueblito['price-food'] = df_pueblito['price-food'].apply(lambda x: (float(x[3:] if x[0] == 'M' else x[1:]) if not isinstance(x, int) else float(x)) if not isinstance(x, float) else x)
```

```python
# Get all "pizza" dataframes
dfPizzasCentro = df_centro[df_centro['name-food'].str.contains("pizza", na=False, case=False)]
dfPizzasAlamos = df_alamos[df_alamos['name-food'].str.contains("pizza", na=False, case=False)]
dfPizzasJuriquilla = df_juriquilla[df_juriquilla['name-food'].str.contains("pizza", na=False, case=False)]
dfPizzasMilenio = df_milenio[df_milenio['name-food'].str.contains("pizza", na=False, case=False)]
dfPizzasRefugio = df_refugio[df_refugio['name-food'].str.contains("pizza", na=False, case=False)]
dfPizzasBalvanera = df_balvanera[df_balvanera['name-food'].str.contains("pizza", na=False, case=False)]
dfPizzasCampanario = df_campanario[df_campanario['name-food'].str.contains("pizza", na=False, case=False)]
dfPizzasCerrito = df_cerritocolorado[df_cerritocolorado['name-food'].str.contains("pizza", na=False, case=False)]
dfPizzasCimatario = df_cimatario[df_cimatario['name-food'].str.contains("pizza", na=False, case=False)]
dfPizzasPueblito = df_pueblito[df_pueblito['name-food'].str.contains("pizza", na=False, case=False)]

dfPizzasCentro.reset_index(drop=True, inplace=True)
dfPizzasAlamos.reset_index(drop=True, inplace=True)
dfPizzasJuriquilla.reset_index(drop=True, inplace=True)
dfPizzasMilenio.reset_index(drop=True, inplace=True)
dfPizzasRefugio.reset_index(drop=True, inplace=True)
dfPizzasBalvanera.reset_index(drop=True, inplace=True)
dfPizzasCampanario.reset_index(drop=True, inplace=True)
dfPizzasCerrito.reset_index(drop=True, inplace=True)
dfPizzasCimatario.reset_index(drop=True, inplace=True)
dfPizzasPueblito.reset_index(drop=True, inplace=True)
```

```python
# Get the whole dataframe
df = pd.concat([dfPizzasCentro, dfPizzasAlamos, dfPizzasJuriquilla, dfPizzasMilenio, dfPizzasRefugio, dfPizzasBalvanera, dfPizzasCampanario, dfPizzasCerrito, dfPizzasCimatario, dfPizzasPueblito])
df.reset_index(drop=True, inplace=True)
```

```python
# Prepare mathplotlib
import matplotlib.pyplot as plt

colonies = ['Centro', 'Alamos', 'Juriquilla', 'Milenio', 'Refugio', 'Balvanera', 'Campanario', 'Cerrito', 'Cimatario', 'Pueblito']
```

```python
# -- FIRST QUESTION --
# ¿Cuál es el menú típico de un restaurante de pizzas?

df_grouped_by_name = df.pivot_table(index = ['name-food'], aggfunc='size')
df_grouped_by_name = df_grouped_by_name.sort_values(ascending = False)

# Get the top 5
top_3 = df_grouped_by_name[0:3]

print('¿Cuál es el menú típico de un restaurante de pizzas? (Top 3)')
for key, data in top_3.items():
    print(f'{key} {data} veces')
```

```python
# -- SECOND QUESTION --
# ¿En qué colonias está la mayor concentración de restaurante de ese tipo?

all_names = []

df_centro_1 = set(dfPizzasCentro.loc[dfPizzasCentro['name-food'] == 'Pizza Hawaiana']['name'])
df_centro_2 = set(dfPizzasCentro.loc[dfPizzasCentro['name-food'] == 'Pizza Vegetariana']['name'])
df_centro_3 = set(dfPizzasCentro.loc[dfPizzasCentro['name-food'] == 'Pizza Mexicana']['name'])
centro_inter = len(df_centro_1.intersection(df_centro_2).intersection(df_centro_3))
all_names.append(list(df_centro_1.intersection(df_centro_2).intersection(df_centro_3)))

df_alamos_1 = set(dfPizzasAlamos.loc[dfPizzasAlamos['name-food'] == 'Pizza Hawaiana']['name'])
df_alamos_2 = set(dfPizzasAlamos.loc[dfPizzasAlamos['name-food'] == 'Pizza Vegetariana']['name'])
df_alamos_3 = set(dfPizzasAlamos.loc[dfPizzasAlamos['name-food'] == 'Pizza Mexicana']['name'])
alamos_inter = len(df_alamos_1.intersection(df_alamos_2).intersection(df_alamos_3))
all_names.append(list(df_alamos_1.intersection(df_alamos_2).intersection(df_alamos_3)))

df_juriquilla_1 = set(dfPizzasJuriquilla.loc[dfPizzasJuriquilla['name-food'] == 'Pizza Hawaiana']['name'])
df_juriquilla_2 = set(dfPizzasJuriquilla.loc[dfPizzasJuriquilla['name-food'] == 'Pizza Vegetariana']['name'])
df_juriquilla_3 = set(dfPizzasJuriquilla.loc[dfPizzasJuriquilla['name-food'] == 'Pizza Mexicana']['name'])
juriquilla_inter = len(df_juriquilla_1.intersection(df_juriquilla_2).intersection(df_juriquilla_3))
all_names.append(list(df_juriquilla_1.intersection(df_juriquilla_2).intersection(df_juriquilla_3)))

df_milenio_1 = set(dfPizzasMilenio.loc[dfPizzasMilenio['name-food'] == 'Pizza Hawaiana']['name'])
df_milenio_2 = set(dfPizzasMilenio.loc[dfPizzasMilenio['name-food'] == 'Pizza Vegetariana']['name'])
df_milenio_3 = set(dfPizzasMilenio.loc[dfPizzasMilenio['name-food'] == 'Pizza Mexicana']['name'])
milenio_inter = len(df_milenio_1.intersection(df_milenio_2).intersection(df_milenio_3))
all_names.append(list(df_milenio_1.intersection(df_milenio_2).intersection(df_milenio_3)))

df_refugio_1 = set(dfPizzasRefugio.loc[dfPizzasRefugio['name-food'] == 'Pizza Hawaiana']['name'])
df_refugio_2 = set(dfPizzasRefugio.loc[dfPizzasRefugio['name-food'] == 'Pizza Vegetariana']['name'])
df_refugio_3 = set(dfPizzasRefugio.loc[dfPizzasRefugio['name-food'] == 'Pizza Mexicana']['name'])
refugio_inter = len(df_refugio_1.intersection(df_refugio_2).intersection(df_refugio_3))
all_names.append(list(df_refugio_1.intersection(df_refugio_2).intersection(df_refugio_3)))

df_balvanera_1 = set(dfPizzasBalvanera.loc[dfPizzasBalvanera['name-food'] == 'Pizza Hawaiana']['name'])
df_balvanera_2 = set(dfPizzasBalvanera.loc[dfPizzasBalvanera['name-food'] == 'Pizza Vegetariana']['name'])
df_balvanera_3 = set(dfPizzasBalvanera.loc[dfPizzasBalvanera['name-food'] == 'Pizza Mexicana']['name'])
balvanera_inter = len(df_balvanera_1.intersection(df_balvanera_2).intersection(df_balvanera_3))
all_names.append(list(df_balvanera_1.intersection(df_balvanera_2).intersection(df_balvanera_3)))

df_campanario_1 = set(dfPizzasCampanario.loc[dfPizzasCampanario['name-food'] == 'Pizza Hawaiana']['name'])
df_campanario_2 = set(dfPizzasCampanario.loc[dfPizzasCampanario['name-food'] == 'Pizza Vegetariana']['name'])
df_campanario_3 = set(dfPizzasCampanario.loc[dfPizzasCampanario['name-food'] == 'Pizza Mexicana']['name'])
campanario_inter = len(df_campanario_1.intersection(df_campanario_2).intersection(df_campanario_3))
all_names.append(list(df_campanario_1.intersection(df_campanario_2).intersection(df_campanario_3)))

df_cerritocolorado_1 = set(dfPizzasCerrito.loc[dfPizzasCerrito['name-food'] == 'Pizza Hawaiana']['name'])
df_cerritocolorado_2 = set(dfPizzasCerrito.loc[dfPizzasCerrito['name-food'] == 'Pizza Vegetariana']['name'])
df_cerritocolorado_3 = set(dfPizzasCerrito.loc[dfPizzasCerrito['name-food'] == 'Pizza Mexicana']['name'])
cerritocolorado_inter = len(df_cerritocolorado_1.intersection(df_cerritocolorado_2).intersection(df_cerritocolorado_3))
all_names.append(list(df_cerritocolorado_1.intersection(df_cerritocolorado_2).intersection(df_cerritocolorado_3)))

df_cimatario_1 = set(dfPizzasCimatario.loc[dfPizzasCimatario['name-food'] == 'Pizza Hawaiana']['name'])
df_cimatario_2 = set(dfPizzasCimatario.loc[dfPizzasCimatario['name-food'] == 'Pizza Vegetariana']['name'])
df_cimatario_3 = set(dfPizzasCimatario.loc[dfPizzasCimatario['name-food'] == 'Pizza Mexicana']['name'])
cimatario_inter = len(df_cimatario_1.intersection(df_cimatario_2).intersection(df_cimatario_3))
all_names.append(list(df_cimatario_1.intersection(df_cimatario_2).intersection(df_cimatario_3)))

df_pueblito_1 = set(dfPizzasPueblito.loc[dfPizzasPueblito['name-food'] == 'Pizza Hawaiana']['name'])
df_pueblito_2 = set(dfPizzasPueblito.loc[dfPizzasPueblito['name-food'] == 'Pizza Vegetariana']['name'])
df_pueblito_3 = set(dfPizzasPueblito.loc[dfPizzasPueblito['name-food'] == 'Pizza Mexicana']['name'])
pueblito_inter = len(df_pueblito_2.intersection(df_pueblito_2).intersection(df_pueblito_3))
all_names.append(list(df_pueblito_2.intersection(df_pueblito_2).intersection(df_pueblito_3)))

have_menu = [centro_inter, alamos_inter, juriquilla_inter, milenio_inter, refugio_inter, balvanera_inter, campanario_inter, cerritocolorado_inter, cimatario_inter, pueblito_inter]

plt.title('¿En qué colonias está la mayor concentración de restaurante de ese tipo?')
plt.xlabel('Número de restaurantes')
plt.ylabel('Colonias')
plt.barh(colonies, have_menu, color=['deepskyblue'])
plt.show()
```

```python
# -- THIRD QUESTION --
# ¿Cómo son los restaurantes de este tipo precios/rating/menú por colonia?

dfColonies = [dfPizzasCentro, dfPizzasAlamos, dfPizzasJuriquilla, dfPizzasMilenio, dfPizzasRefugio, dfPizzasBalvanera, dfPizzasCampanario, dfPizzasCerrito, dfPizzasCimatario, dfPizzasPueblito]
graph_min = []
graph_max = []
graph_rating = []

for i in range(len(dfColonies)):
    dfColonies[i].reset_index(drop=True, inplace=True)
    i_menu = dfColonies[i].pivot_table(index = ['name-food'], aggfunc = 'size').sort_values(ascending = False)[0:3]
    i_lower = dfColonies[i].loc[dfColonies[i]['price-food'] == dfColonies[i]['price-food'].min()]
    i_higher = dfColonies[i].loc[dfColonies[i]['price-food'] == dfColonies[i]['price-food'].max()]
    i_best = dfColonies[i].loc[dfColonies[i]['rating'] == dfColonies[i]['rating'].max()]
    i_mean = dfColonies[i]['price-food'].mean()
    
    graph_min.append(dfColonies[i]['price-food'].min())
    graph_max.append(dfColonies[i]['price-food'].max())
    graph_rating.append(dfColonies[i]['rating'].max())

    print(f'\n\n-- {colonies[i]} --')
    print(f"\nPrecio promedio: ${i_mean}")
    print('\nMenu (Top 3):')
    for key, data in i_menu.items():
        print(f'{key} {data} veces')
    print('\nPrecios más bajo:')
    for index, row in i_lower.iterrows():
        print(f"{row['name-food']} con ${row['price-food']}")
    print('\nPrecios más altos:')
    for index, row in i_higher.iterrows():
        print(f"{row['name-food']} con ${row['price-food']}")
    print('\nRating más altos:')
    for index, row in i_best.drop_duplicates(subset=['name']).iterrows():
        print(f"{row['name']} con un rating de {row['rating']}")
```

```python
# -- FORTH QUESTION --
# Top 10 de esos restaurante en la ciudad

all_rates = None

top_restaurants = []
top_rating_values = []

for i in range(len(dfColonies)):
    for j in range(len(all_names[i])):
        preDf = dfColonies[i].loc[dfColonies[i]['name'] == all_names[i][j]]
        all_rates_temp = pd.concat([all_rates, preDf], ignore_index=True)
        all_rates = all_rates_temp

all_rates = all_rates.sort_values(by=['rating'], ascending=False).drop_duplicates(subset=['name'])[0:10]
top = 1
print('-- Top 10 de esos restaurante en la ciudad --')
for index, row in all_rates.iterrows():
    print(f"Top #{top}: {row['name']} con un rating de {row['rating']}")
    top_restaurants.append(row['name'])
    top_rating_values.append(row['rating'])
    top += 1

plt.title('Top 10 de esos restaurante en la ciudad')
plt.xlabel('Rating')
plt.ylabel('Restaurantes')
plt.barh(top_restaurants, top_rating_values, color=['deepskyblue'])
plt.show()
```

```python
# -- FIRTH QUESTION --
# Mejor colonia en Qro para poner un restaurante de ese tipo

print('\nLa mejor colonia para poner un restaurante es Balvanera, porque tiene los precios más bajos, un buen rating y aparte tiene el menor número de restaurantes con el menú típico, por lo que las oportunidad son mayores allá')

meanPizzaVal = df['price-food'].mean()
meanPizza = df.loc[df['price-food'] == meanPizzaVal]
print(f'\n\nEl precio promedio de una pizza es de ${meanPizzaVal}:')
print(meanPizza.loc[:,['name','name-food','price-food']])

medianPizza = df.loc[df['price-food'] == df['price-food'].median()]
print('\n\nLa mediana de las pizzas es:')
print(medianPizza.loc[:,['name','name-food','price-food']])

modePizza = df.loc[df['price-food'] == df['price-food'].mode()[0]]
print('\n\nLa moda de las pizzas es:')
print(modePizza.loc[:,['name','name-food','price-food']])

stdPizza = df['price-food'].std()
varPizza = df['price-food'].var()

print(f'\n\nLa varianza de la pizza es {varPizza}')
print(f'La desviación estándar de la pizza es ${stdPizza}')

minPizza = df.loc[df['price-food'] == df['price-food'].min()]
print('\n\nLas pizzas con el precio más bajo son:')
print(minPizza.loc[:,['name','name-food','price-food']])

maxPizza = df.loc[df['price-food'] == df['price-food'].max()]
print('\n\nLas pizzas con el precio más elevado son:')
print(maxPizza.loc[:,['name','name-food','price-food']])

plt.title('Precio más bajo')
plt.xlabel('Precio')
plt.ylabel('Restaurantes')
plt.barh(colonies, graph_min, color=['deepskyblue'])
plt.show()

plt.title('Precio más alto')
plt.xlabel('Precio')
plt.ylabel('Restaurantes')
plt.barh(colonies, graph_max, color=['deepskyblue'])
plt.show()

plt.title('Mayor rating')
plt.xlabel('Rating')
plt.ylabel('Restaurantes')
plt.barh(colonies, graph_rating, color=['deepskyblue'])
plt.show()
```
